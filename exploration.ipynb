```python
import pandas as pd
```

```python
df = pd.read_csv('ncts_with_original_df.csv')
# Extract only the first zipcode from the Locations column
# This assumes zipcodes are 5-digit numbers in the format shown in the example
import re

# Function to extract the first zipcode from location strings
def extract_first_zipcode(location_str):
    if pd.isna(location_str):
        return None
    # Pattern to match 5-digit zipcodes
    pattern = r'\b\d{5}\b'
    matches = re.findall(pattern, location_str)
    # Return only the first zipcode if found, otherwise None
    return matches[0] if matches else None

# Apply the function to each row to get the first zipcode
df['first_zipcode'] = df.Locations.apply(extract_first_zipcode)

# Display an example of the extraction
example = df[df['first_zipcode'].notna()].iloc[0] if any(df['first_zipcode'].notna()) else None

if example is not None:
    print(f"Example of first zipcode extraction:")
    print(f"NCT ID: {example.get('NCT ID', 'N/A')}")
    print(f"Locations: {example.Locations}")
    print(f"First zipcode: {example.first_zipcode}")
else:
    print("No zipcodes found in the dataset.")

# Extract all unique first zipcodes for reference
zipcodes = df['first_zipcode'].dropna().unique()
zipcodes
```
